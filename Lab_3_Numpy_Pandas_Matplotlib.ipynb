## 4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia a-b).

```python
import numpy as np

# Definir los arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia elemento a elemento entre a y b
diferencia = a - b

# Calcular la media aritmética de la diferencia
media_diferencia = np.mean(diferencia)

# Imprimir el resultado
print("La media aritmética de la diferencia (a-b) es:", media_diferencia)
```

## 2) Realice una gráfica de barras sobre la polbación de cada ciudad, ordenela de menor a mayo y responda:
## a. ¿Cuál es la segunda ciudad más poblada?
## b. ¿Qué posición ocupa Granada en el ranking de 
## las ciudades más pobladas?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV
dataset = pd.read_csv("NucleosPoblacion.csv", delimiter=';')

# Ordenar los datos según la población de menor a mayor
dataset_ordenado = dataset.sort_values(by='Poblacion')

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
plt.bar(dataset_ordenado['Provincia'], dataset_ordenado['Poblacion'], color='skyblue')
plt.xticks(rotation=90)  # Rotar etiquetas del eje x para una mejor visualización
plt.xlabel('Ciudad')
plt.ylabel('Población')
plt.title('Población de cada ciudad')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar la gráfica
plt.show()

# Respuestas a las preguntas
# a. Segunda ciudad más poblada
segunda_ciudad = dataset_ordenado.iloc[-2]['Provincia']
segunda_ciudad_poblacion = dataset_ordenado.iloc[-2]['Poblacion']

# b. Posición de Granada en el ranking de las ciudades más pobladas
posicion_granada = dataset_ordenado[dataset_ordenado['Provincia'] == 'Granada'].index[0] + 1

print("a. La segunda ciudad más poblada es:", segunda_ciudad, "con una población de", segunda_ciudad_poblacion)
print("b. Granada ocupa la posición", posicion_granada, "en el ranking de las ciudades más pobladas.")
```

## 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV
dataset = pd.read_csv("NucleosPoblacion.csv", delimiter=';')

# Obtener una lista de provincias únicas
provincias = dataset['Provincia'].unique()

# Crear subgráficos para cada provincia
fig, axs = plt.subplots(len(provincias), figsize=(10, 6*len(provincias)), sharex=True)

# Iterar sobre cada provincia y crear un histograma para la población de sus municipios
for i, provincia in enumerate(provincias):
    # Filtrar los datos para la provincia actual
    datos_provincia = dataset[dataset['Provincia'] == provincia]
    # Crear el histograma
    axs[i].hist(datos_provincia['Poblacion'], bins=20, color='skyblue', edgecolor='black')
    axs[i].set_title(f'Histograma de población en {provincia}')
    axs[i].set_xlabel('Población')
    axs[i].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

## 7) Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
import pandas as pd

# Cargar los datos del archivo CSV
dataset = pd.read_csv("NucleosPoblacion.csv", delimiter=';')

# Seleccionar al azar cincuenta municipios diferentes
municipios_aleatorios = dataset['Municipio'].sample(n=50, replace=False)

# Filtrar el dataset para incluir solo los municipios seleccionados al azar
municipios_seleccionados = dataset[dataset['Municipio'].isin(municipios_aleatorios)]

# Calcular el promedio y la desviación estándar de la población de estos cincuenta municipios
promedio_poblacion = municipios_seleccionados['Poblacion'].mean()
desviacion_estandar_poblacion = municipios_seleccionados['Poblacion'].std()

print("Promedio de la población de los cincuenta municipios seleccionados al azar:", promedio_poblacion)
print("Desviación estándar de la población de los cincuenta municipios seleccionados al azar:", desviacion_estandar_poblacion)
```

## 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
import pandas as pd
from geopy.distance import geodesic

# Coordenadas de Madrid (latitud y longitud)
madrid_coords = (40.4168, -3.7038)

# Cargar los datos del archivo CSV
dataset = pd.read_csv("NucleosPoblacion.csv", delimiter=';')

# Función para calcular la distancia entre Madrid y un municipio
def calcular_distancia_madrid(row):
    municipio_coords = (row['Y'], row['X'])  # Coordenadas del municipio
    distancia = geodesic(madrid_coords, municipio_coords).kilometers
    return distancia

# Calcular la distancia entre Madrid y cada municipio
dataset['Distancia_Madrid'] = dataset.apply(calcular_distancia_madrid, axis=1)

# Encontrar el municipio más cercano y más lejano a Madrid
municipio_mas_cercano = dataset.loc[dataset['Distancia_Madrid'].idxmin()]
municipio_mas_lejano = dataset.loc[dataset['Distancia_Madrid'].idxmax()]

print("Municipio más cercano a Madrid:", municipio_mas_cercano['Municipio'])
print("Distancia al municipio más cercano:", municipio_mas_cercano['Distancia_Madrid'], "kilómetros")
print("Municipio más lejano a Madrid:", municipio_mas_lejano['Municipio'])
print("Distancia al municipio más lejano:", municipio_mas_lejano['Distancia_Madrid'], "kilómetros")
```
